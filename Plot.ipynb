## Input Files
For each set of predictions, we need the production rate for all charm hadron ($D^+$, $D^0$, $D_s^+$, $\Lambda_C^+$ as well as their antiparticles). The production rates should be provided as table of double differential cross section $d\sigma/(dp_T dy)$ in bins of transverse momentum $p_T$ (with $p_{T,min}=0.125$ GeV,  $p_{T,max}=29.725$ GeV,  $\Delta p_{T}=0.25$ GeV) and rapidity $y$ (with $y_{min}=0.25$,  $y_{max}=9.75$,  $\Delta y=0.5$). 

The files are saved in the `files` directory. Currently we the following predictions:

- MC Generator predictions (obtained by Felix)
    - `SIBYLL`, (version 2.3d) see [1806.04140](https://arxiv.org/abs/1806.04140)
    - `DPMJET`, (version 3.2019) see [GitHub](https://github.com/DPMJET/DPMJET)
    - `Pythia8` (SoftQCD process, default setting using Monash tune), see [1404.5630](https://arxiv.org/abs/1404.5630)
    - `Pythia8_BLC` (SoftQCD process, using string formation beyond leading color), see [1505.01681](https://arxiv.org/abs/1505.01681)
    - `Pythia8_Hard` (HardQCD process, default setting using Monash tune)
    
    

- Predictions using kT factorization (obtained by Anna, Atri, Ina, Felix)  
    - `kt_linear`, (no saturation, CT14nlo for hard gluon, Fragmentation ala Pythia8 BLC)
    - `kt_saturation`, (with saturation, CT14nlo for hard gluon, Fragmentation ala Pythia8 BLC)
